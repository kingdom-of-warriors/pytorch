```python
# 准备工作
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 目标函数和它的导数
def fun(x):
    return 2 * (x ** 4) - 9 * (x ** 3) - 21 * (x ** 2) + 88 * x + 48

def df(x):
    return 8 * (x**3) - 27 * (x**2) - 42 * x + 88

def plot_fun():
    x = np.arange(-4,6,0.1)
    y = fun(x)
    plt.plot(x,y,label = 'function')
    plt.legend()
    plt.show()

plot_fun()
lr = 0.001
EPOCHS = 10
x_list = [5]
x = x_list[0]
y_list = [fun(x)]
```

```python
# 先是没有momentum的梯度下降
for epoch in range(EPOCHS):
    x-=df(x)*lr
    x_list.append(x)
    y_list.append(fun(x))

plt.scatter(x_list,y_list,s = 15,c = 'r')
plot_fun()

#可以看出没有越过局部最低点
```

```python
# 有momentum的梯度下降

EPOCHS = 100
lr = 0.009
x_list = [5]
x = x_list[0]
y_list = [fun(x)]
v_list = [0]
v = v_list[0]
r = 0.9
for epoch in range(EPOCHS):
    v = r * v + df(x)
    x = x - lr * v
    x_list.append(x)
    y_list.append(fun(x))

plt.scatter(x_list,y_list,s = 15,c = 'r')
plot_fun()

# 这个越过了局部最低，找到全局最低
```
